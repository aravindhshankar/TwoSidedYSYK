```python
import numpy as np 
from scipy.fft import fft,ifft,fftshift,fftfreq
from matplotlib import pyplot as plt
```

```python
t = np.linspace(0,5,100)
x = np.exp(-0.5*(t**2))
y = fft(x)
freqs = fftfreq(t.size,d=t[1]-t[0])
```

```python
#plt.plot(np.abs(y),'r')
plt.plot(fftshift(freqs),np.abs(fftshift(y)))
```

```python
plt.plot(t,ifft(y))
plt.plot(t,ifft(fftshift(y)),'r--')
#plt.plot(t,x,'k.')
```
